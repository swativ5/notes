```python
import numpy as np

def build_sample_image_color(colors):
    """Builds a sample image with 50x50 regions of different colors"""

    # Initialize result with the first 50x50 region with black color
    result = np.zeros((50, 50, 3), dtype="uint8")
 
    for color in colors:
             img = np.ones((50, 50, 3), dtype="uint8") * color
             result = np.concatenate((result, img), axis=1)

    return result
colors = np.array([(255,0,255),(255,0,0),(255,255,255),(0,255,0),(0,0,255),(255,255,0),(0,255,255)], dtype="uint8")
color_img = build_sample_image_color(colors)

#display image using matplotlib
import matplotlib.pyplot as plt
plt.imshow(color_img)
```

```python
import matplotlib.pyplot as plt
import numpy as np


black = (0,0,0)
pink = (255,0,255)
blue = (255,0,0)
white = (255,255,255)
green = (0,255,0)
red = (0,0,255)
cyan = (0,255,255)
yellow = (255,255,0)
row = np.array([black, pink, blue, white, green, red, cyan, yellow], dtype="uint8")

def build_sample_image2(row):
    # np array with all elemets as white
    result = np.ones((450, 400, 3), dtype="uint8") * 255

    # first row
    for i in range(8):
        color = row[i]
        img = np.ones((50, 50, 3), dtype="uint8") * color
        result[0:50, i*50:(i+1)*50] = img

    # first column
    for j in range(8):
        color = row[j]
        img = np.ones((50, 50, 3), dtype="uint8") * color
        result[j*50:(j+1)*50, 0:50] = img

    # last column
    for k in range(8):
        color = row[k]
        img = np.ones((50, 50, 3), dtype="uint8") * color
        result[(k+1)*50:(k+2)*50, 350:400] = img

    # last row
    for l in range(8):
        color = row[l]
        img = np.ones((50, 50, 3), dtype="uint8") * color
        result[400:450, l*50:(l+1)*50] = img
    plt.imshow(result)
    return result

result = build_sample_image2(row)
# save image in directory
plt.imsave("sample_image2.png", result)
```

```python
# read image opencv
import cv2

# read the image "sample_image2.png"
img = cv2.imread("sample_image2.png")

# convert to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# show using matplotlib
plt.imshow(gray_img, cmap='gray')

# flip the image vertically
flipped_img = cv2.flip(img, 0)
```
