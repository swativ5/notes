```python
import math
import matplotlib.pyplot as plt

t = 0
delta = 0.01 
theta_t = math.pi / 4  
w_t = 0  
g = 10
l = 10

time_values = []
theta_values = []

for i in range(3000):
    time_values.append(t)
    theta_values.append(theta_t)
    
    x = theta_t - ((theta_t ** 3) / 6)
    
    w_t = w_t - ((g / l) * x * delta)
    theta_t = theta_t + w_t * delta - (g / l) * x * (delta ** 2) / 2
    
    t += delta


plt.figure(figsize=(10, 6))
plt.plot(time_values, theta_values, 'b-', linewidth=2)
plt.xlabel('Time (s)')
plt.ylabel('Theta (radians)')
plt.grid(True)
plt.show()
```

```python
# Find peaks
peaks = []
i = 0

while i < len(theta_values):
    # First value in current segment is a peak
    peak_val = theta_values[i]
    peak_time = time_values[i]
    peaks.append((peak_time, peak_val))
    print(f"Peak found at time {peak_time:.3f}s: theta = {peak_val:.4f} radians")
    
    # Cut out values until theta goes to 0 (or changes sign)
    i += 1
    while i < len(theta_values) and theta_values[i] * peak_val > 0:
        i += 1
    
    # Skip the zero crossing
    while i < len(theta_values) and abs(theta_values[i]) < 0.01:
        i += 1
    
    # Find the max of the rest of values in this swing
    if i < len(theta_values):
        max_val = theta_values[i]
        max_idx = i
        j = i + 1
        
        # Keep looking for max until we hit zero again
        while j < len(theta_values) and theta_values[j] * theta_values[i] > 0:
            if abs(theta_values[j]) > abs(max_val):
                max_val = theta_values[j]
                max_idx = j
            j += 1
        
        i = max_idx

print(f"\nTotal peaks found: {len(peaks)}")
```
