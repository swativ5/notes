Find the polynomial of degree two that passes through three points: (0, 1), (-1, 2) and (1, 3)

Points:
- t0, ft0 = 0, 1
- t1, ft1 = -1, 2
- t2, ft2 = 1, 3

Polynomial of degree two:
- Pt = c0 + c1.t + c2.t^2

Substituting the points into the polynomial:
- Pt0 = c0 + c1.t0 + c2.t0^2 = c0 + 0 + 0 = c0 = ft0 = 1 -> **c0 = 1**
- Pt1 = c0 + c1.t1 + c2.t1^2 = 1 - c1 + c2 = ft1 = 2 -> -c1 + c2 = 1 -> c1 = c2 - 1
- Pt2 = c0 + c1.t2 + c2.t2^2 = 1 + c1 + c2 = ft2 = 3 -> c1 + c2 = 2 -> c1 = 2 - c2
- 2 - c2 = c2 - 1-> 2c2 = 3 -> **c2 = 3/2** -> c1 = 2 - 3/2 = 1/2 -> **c1 = 1/2**

````python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linspace

def P(t):
    c0 = 1
    c2 = 3/2
    c1 = 1/2
    return c0 + c1*t + c2*t**2

t = linspace(-1, 1, 100)
plt.plot(t, P(t), label='P(t)')
plt.scatter([-1, 0, 1], [2, 1, 3], color='red', label='Given Points')
plt.show()
````

````python
def createA(x): #aka vandermonde matrix
    n = len(x)
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            A[i, j] = x[i]**j
    return A

def createb(y):
    b = np.array([y[i] for i in range(len(y))])
    return b

x = [0, -1, 1]
y = [1, 2, 3]
A = createA(x)
b = createb(y)
coefficients = np.linalg.solve(A, b)

def P(x, coefficients):
    sum = 0
    idx = 0
    for i in coefficients:
        sum += i * x**idx
        idx += 1
    return sum

t = linspace(-1, 1, 100)
plt.plot(t, P(t, coefficients), label='P(t)')
plt.scatter([-1, 0, 1], [2, 1, 3], color='red', label='Given Points')
plt.show()
````
